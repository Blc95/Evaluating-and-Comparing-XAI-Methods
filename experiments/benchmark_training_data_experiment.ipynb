# Computing Correlation

```python
def avg_abs_correlation(X: pd.DataFrame) -> float:
    """
    Compute average absolute pairwise correlation between features.
    """
    corr = X.corr()
    abs_corr = corr.abs()

    # Remove diagonal
    upper = abs_corr.where(np.triu(np.ones(abs_corr.shape), k=1).astype(bool))

    return upper.stack().mean()
```

# Define base model that will be trained on different amount of training data

```python
MODEL_RANDOM_STATE = 20

# Define baseline models, scaling the data where needed

# Base Logistic Regression model
base_lm = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=2000,
        random_state=MODEL_RANDOM_STATE))
])

# Base Random Forest model
base_rf = Pipeline([
    ("model", RandomForestClassifier(
        n_estimators=500,
        random_state=MODEL_RANDOM_STATE))
])

# Base Gradient Boosting model
base_gb = Pipeline([
    ("model", GradientBoostingClassifier(
        n_estimators=500,
        random_state=MODEL_RANDOM_STATE))
])

# Base MLP model
base_nn = Pipeline([
    ("scaler", StandardScaler()),
    ("model", MLPClassifier(
        hidden_layer_sizes=(16,),
        solver="adam",
        alpha=1e-4, learning_rate_init=1e-3,
        max_iter=2000, early_stopping=True,
        n_iter_no_change=20,
        random_state=MODEL_RANDOM_STATE
    ))
])

```

```python
SEED = 1

tiers = {
    "low": 0.10,
    "med": 0.50,
    "high": 1.00
}

def make_subset(X_full, y_full, frac, seed):
    """Return a stratified subset of size 'frac'"""
    if frac >= 0.999:
        return X_full, y_full
    X_sub, _, y_sub, _ = train_test_split(
        X_full, y_full, train_size=frac, stratify=y_full, random_state=seed
    )
    return X_sub, y_sub
```

# Training the Models

```python
rows = []
models_store = {}

models_performance = {}

# Change train set here
X_train_initial = X_bankn_initial_train
y_train_initial = y_bankn_initial_train

# change test set here
X_test = X_bankn_test
y_test = y_bankn_test

# Change name of dataset here
dataset = "banknote"


# Change model here
model = base_nn
# change name of model here
model_name = "nn"


for tier_name, frac in tiers.items():
        # Sample the training subset for this tier/seed
        X_train_sub, y_train_sub = make_subset(X_train_initial, y_train_initial, frac, SEED)

        # Train the same model on this subset
        model = clone(model) ### CHANGE MODEL HERE ###
        model.fit(X_train_sub, y_train_sub)

        # Evaluate on the test set
        preds = model.predict(X_test)
        acc = accuracy_score(y_test, preds)
        f1_pos = f1_score(y_test, preds, average="binary", pos_label=1)

        # Probabilities of predictions
        proba = model.predict_proba(X_test)
        y_score = proba[:, 1]

        # Average precision
        ap = average_precision_score(y_test, y_score)

        # ROC
        roc = roc_auc_score(y_test, y_score)

        # Confidence in model choice
        confidences = np.max(proba, axis=1)

        # Confidecen for right/wrong predictions
        y_true = np.asarray(y_test)
        correct = (preds == y_true)
        mean_confidence_correct = np.mean(confidences[correct])
        mean_confidence_wrong   = np.mean(confidences[~correct])

        # Mean confidence across predictions
        mean_confidence = np.mean(confidences)


        rows.append({
            "model": model_name,
            "dataset": dataset,
            "tier": tier_name,
            "n_train": len(X_train_sub),
            "accuracy": acc,
            "f1_pos": f1_pos,
            "ap": ap,
            "roc_auc": roc,
            "mean_confidence": mean_confidence,
            "mean_confidence_correct": mean_confidence_correct,
            "mean_confidence_wrong": mean_confidence_wrong
        })
        models_store[(dataset, model_name, tier_name)] = model

        models_performance[(dataset, model_name, tier_name)] = ({
            "n_train": len(X_train_sub),
            "accuracy": acc,
            "f1_pos": f1_pos,
            "ap": ap,
            "roc_auc": roc,
            "mean_confidence": mean_confidence,
            "mean_confidence_correct": mean_confidence_correct,
            "mean_confidence_wrong": mean_confidence_wrong
          })

results = pd.DataFrame(rows).sort_values(["tier"])
#print(results)
#print(results.groupby("tier")[["accuracy","f1_pos","ap","roc_auc", "mean_confidence", "mean_confidence_correct", "mean_confidence_wrong"]].agg(["mean","std","min","max"]))
```

```python
import warnings, contextlib, os, pickle

metrics = {}
filenames = []

INSTANCES_PER_CLASS = 40
global_seed = np.random.default_rng(123)


def stratified_indices(y, per_class=10, s=None):
    '''Function for finding X instances per class'''
    y = y.ravel()
    idxs = [] # Empty list for storing indexes of instances from each class
    for c in np.unique(y): # Loop over each unique class
        cls_idx = np.where(y == c)[0] # Get indicies all all instances belonging to c
        chosen = s.choice(cls_idx, size=per_class, replace=False) # Randomly choose per_class indicies from the class index
        idxs.append(chosen) # append the chosen classes to the list of instances chosen
    idxs = np.concatenate(idxs) # concatenate the list of array into one list
    s.shuffle(idxs) # Shuffle instances
    return idxs

### CHANGE STUFF HERE ###

# Change train set here
X_train_initial = X_bankn_initial_train
y_train_initial = y_bankn_initial_train

# Change test set here
X_test = X_bankn_test
y_test = y_bankn_test

# Change name of dataset here
dataset = "banknote"

# Change method here
method = "lime"

# Change model here
model_name = "nn"
######################

test_idx = stratified_indices(y_test, per_class=INSTANCES_PER_CLASS, s=global_seed)

X_test = X_test.iloc[test_idx].values
y_test = y_test.iloc[test_idx].values



# Define traning fractions
tiers = {"low": 0.10, "med": 0.50, "high": 1.00}

# Loop through tiers and seeds
for tier_name, frac in tiers.items():

        key = (dataset, model_name, tier_name)
        print("Running metrics for:", tier_name)

        # Get the trained model
        model = models_store[key]

        # Recreate the training subset used for this model (same frac + seed)
        # This is only to derive a background set for explainers.
        X_train_sub, y_train_sub = make_subset(X_train_initial, y_train_initial, frac, SEED)

        # Background
        X_tr_bg = X_train_sub.values

        # Build the metric runner
        em = ExplanationMetricsTabular(
            model=model,
            explainer_name=method,
            X_train=shap.sample(X_tr_bg, 200),  # background from TRAIN SUBSET ONLY
            X_test=X_test,
            y_test=y_test,
            features=feature_names,
            task="binary-classification"
        )

        # Compute metrics and quiet spammy warnings
        with warnings.catch_warnings():
            with open(os.devnull, "w") as devnull, \
                 contextlib.redirect_stdout(devnull), \
                 contextlib.redirect_stderr(devnull):
                result = em.calculate_metrics()

        metrics[key] = result
        del em

        filename = f"{dataset}_{model_name}_metrics_{tier_name}_{method}_{frac}.pkl"
        with open(filename, "wb") as f:
            pickle.dump({key: result}, f)
        filenames.append(filename)
```
